```python
#Open webdriver 
driver = webdriver.Chrome(path+"chromedriver.exe")
driver.get('https://shop.villamarket.com/catalog/Confectionery%20&%20Snacks/All/All')
```

```python
food = driver.find_elements(By.TAG_NAME,"a")
img = driver.find_elements(By.TAG_NAME,"img")

food[0].find_element(By.CSS_SELECTOR,"div[class*='name-text']").get_attribute("textContent")
img[0].get_attribute("src").replace("https://flagsapi.com/","").replace("/flat/64.png","")
```

```python
def extract_df(element):

    def food(element):
        food = element.find_elements(By.TAG_NAME,"a")
        return food
    
    def img(element):
        img = element.find_elements(By.TAG_NAME,"img")
        return img

    def extract_data(element1,element2):

        name = element1.find_element(By.CSS_SELECTOR,"div[class*='name-text']").get_attribute("textContent")
        sub_name = element1.find_element(By.CSS_SELECTOR,"div[class*='sub-name-text']").get_attribute("textContent")
        price = element1.find_element(By.CSS_SELECTOR,"span[class*='special-price']").get_attribute("textContent")
        image = element2.get_attribute("src")
        full_name = name+sub_name
        full_name = full_name.strip()
        return {
            "name": full_name,
            "price":price.replace("฿",""),
            "Country":image.replace("https://flagsapi.com/","").replace("/flat/64.png","")
        }
    
    food = food(element)
    img = img(element)
    data = []
    for f,i in zip(food,img):
        extracted_data = extract_data(f,i)
        data.append(extracted_data)
    df = pd.DataFrame(data)

    return(df)
```

```python
df = extract_df(driver)
display(df)
```
